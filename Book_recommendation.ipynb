## Popualarity based Model

```python
r_n = rating.merge(books,on='ISBN')
r_n.head()
```

```python
r_n.groupby('Book-Title').count()
```

```python
r_n.groupby('Book-Title').count()['Book-Rating']   # its a series
```

```python
no_r_per_book = r_n.groupby('Book-Title').count()['Book-Rating'].reset_index()  # converting the above series into df
no_r_per_book.rename(columns={'Book-Rating':'num_rating'},inplace=True)
no_r_per_book.head()
```

```python
avg_r_per_book = r_n.groupby('Book-Title').mean()['Book-Rating'].reset_index()  # avg-mean no. of rating per book
avg_r_per_book.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_r_per_book.head()
```

```python
popular_df = avg_r_per_book.merge(no_r_per_book, on='Book-Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
popular_df.head(14)
```

```python
popular_df = popular_df.merge(r_n,on='Book-Title')
popular_df.head(14)
```
